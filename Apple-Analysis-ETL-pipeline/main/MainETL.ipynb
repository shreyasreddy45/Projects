```python
class FirstWorkFlow:

    def __init__(self):
        pass

    def runner(self):

        print("=" * 50)
        print("FIRST WORKFLOW STARTED")
        print("Goal: Customers who bought AirPods after iPhone")
        print("=" * 50)

        # Step 1: Extract
        print("\n[STEP 1] Extracting data...")
        inputDFs = (
            AirpodsAfterIphoneExtractor()
                .extract()
        )
        print("✓ Extraction complete")

        # Step 2: Transform
        print("\n[STEP 2] Transforming data...")
        firstTransformedDF = (
            AirpodsAfterIphoneTransformer()
                .transform(inputDFs)
        )
        print("✓ Transformation complete")
        print("\nFirst Workflow Output:")
        display(firstTransformedDF)

        # Step 3: Load
        print("\n[STEP 3] Loading data...")
        AirPodsAfterIphoneLoader(
            firstTransformedDF
        ).sink()
        print("✓ Data saved to default.airpodsAfterIphone")
        print("\n" + "=" * 50)
        print("FIRST WORKFLOW COMPLETED")
        print("=" * 50)


class SecondWorkFlow:

    def __init__(self):
        pass

    def runner(self):

        print("=" * 50)
        print("SECOND WORKFLOW STARTED")
        print("Goal: Customers who bought ONLY iPhone and AirPods")
        print("=" * 50)

        # Step 1: Extract
        print("\n[STEP 1] Extracting data...")
        inputDFs = (
            AirpodsAfterIphoneExtractor()
                .extract()
        )
        print("✓ Extraction complete")

        # Step 2: Transform
        print("\n[STEP 2] Transforming data...")
        onlyAirpodsAndIphoneDF = (
            OnlyAirpodsAndIphone()
                .transform(inputDFs)
        )
        print("✓ Transformation complete")
        print("\nSecond Workflow Output:")
        display(onlyAirpodsAndIphoneDF)

        # Step 3: Load
        print("\n[STEP 3] Loading data...")
        OnlyAirpodsAndIPhoneLoader(
            onlyAirpodsAndIphoneDF
        ).sink()
        print("✓ Data saved to default.onlyAirPodsAndIphone")
        print("\n" + "=" * 50)
        print("SECOND WORKFLOW COMPLETED")
        print("=" * 50)


class ThirdWorkFlow:

    def __init__(self):
        pass

    def runner(self):

        print("=" * 50)
        print("THIRD WORKFLOW STARTED")
        print("Goal: Customers who bought iPhone but NEVER AirPods")
        print("=" * 50)

        # Step 1: Extract
        print("\n[STEP 1] Extracting data...")
        inputDFs = (
            AirpodsAfterIphoneExtractor()
                .extract()
        )
        print("✓ Extraction complete")

        # Step 2: Transform
        print("\n[STEP 2] Transforming data...")
        iphoneWithoutAirpodsDF = (
            IphoneWithoutAirpods()
                .transform(inputDFs)
        )
        print("✓ Transformation complete")
        print("\nThird Workflow Output:")
        display(iphoneWithoutAirpodsDF)

        # Step 3: Load
        print("\n[STEP 3] Loading data...")
        IphoneWithoutAirpodsLoader(
            iphoneWithoutAirpodsDF
        ).sink()
        print("✓ Data saved to default.iphoneWithoutAirpods")
        print("\n" + "=" * 50)
        print("THIRD WORKFLOW COMPLETED")
        print("=" * 50)


class WorkFlowRunner:

    def __init__(self, names):
        self.names = names

    def runner(self):

        print("\n" + "*" * 50)
        print("ETL PIPELINE STARTED")
        print(f"Running workflows: {self.names}")
        print("*" * 50 + "\n")

        for name in self.names:

            if name == "FirstWorkFlow":
                FirstWorkFlow().runner()

            elif name == "SecondWorkFlow":
                SecondWorkFlow().runner()

            elif name == "ThirdWorkFlow":
                ThirdWorkFlow().runner()

            else:
                raise ValueError(
                    f"Invalid Workflow: {name}"
                )

        print("\n" + "*" * 50)
        print("ETL PIPELINE COMPLETED SUCCESSFULLY")
        print("*" * 50)


# EXECUTION — all three workflows run together
names = ["FirstWorkFlow", "SecondWorkFlow", "ThirdWorkFlow"]

WorkFlowRunner(names).runner()
```
